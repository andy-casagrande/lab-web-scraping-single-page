```python
url = 'https://www.billboard.com/charts/hot-100'
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

artist_name = []
for artist in soup.find_all('span', class_ = 'lrv-u-font-size-14@mobile-max'):
    artist_name.append(artist.getText().strip())
artist_name
```

```python
song_name = []
for song in soup.find_all('h3', class_ = 'c-title'):
    song_name.append(song.getText().strip())
song_name = song_name[6:403:4]
song_name
```

```python
billboard_hot_100 = pd.DataFrame({
    'Artist': artist_name,
    'Song': song_name
})
billboard_hot_100
```
